Método de Euler

```python
# Importar as bibliotecas
import numpy
import matplotlib.pyplot as plt
```

```python
# Definição do Campo
def F(x,y):
    return -3*y
```

```python
# Inicialização das Variáveis
x = numpy.linspace(-1,3,num=65)
y = numpy.zeros_like(x)
```

```python
# Condição inicial
y[0] = 1
```

```python
# Método de Euler
for i in range(len(x)-1):
    y[i+1] = y[i] + F(x[i],y[i])*(x[i+1]-x[i])
```

```python
# Gráfico da Aproximação
fig = plt.figure()
plt.plot(x,y, label='Método de Euler')
plt.plot(x,numpy.exp(-3*(x+1)), label='Solução')
plt.legend()
plt.show()
```
